```python
#import dependencies
import pandas as pd
```

```python
#set csv paths
csv_path = "Median_Salary_County.csv"
csv_path1 = "Median House Price by County.csv"
```

```python
#create initial DataFrame for median incomes
salary_median_df = pd.read_csv(csv_path)
salary_median_df = salary_median_df[["Area_name", "Median_Household_Income_2020"]]
salary_median_df = salary_median_df.dropna(how="any")
salary_median_df['Area_name'] = salary_median_df['Area_name'].str.split(',').str[0]
salary_median_df

# salary_median_df['Median_Household_Income_2020'] = salary_median_df['Median_Household_Income_2020'].str.split(',').str[0] + salary_median_df['Median_Household_Income_2020'].str.split(',').str[1]
# salary_median_df = salary_median_df.rename(columns={"Area_name": "RegionName"})
# salary_median_df
```

```python
salary_median_df['Median_Household_Income_2020'] = salary_median_df['Median_Household_Income_2020'].str.split(',').str[0] + salary_median_df['Median_Household_Income_2020'].str.split(',').str[1]
salary_median_df = salary_median_df.rename(columns={"Area_name": "RegionName"})
salary_median_df
```

```python
#create DataFrame for median home values
house_median_df = pd.read_csv(csv_path1)
house_median_df = house_median_df[["RegionName", "State", "1/31/2020"]]

#drop null values
house_median_df = house_median_df.dropna(how="any")
house_median_df
```

```python
#merge previous DataFrames
merge_df = pd.merge(salary_median_df, house_median_df, on="RegionName" )

#rename columns of merged DataFrame
merge_df=merge_df.rename(columns={
    "RegionName":"County Name",
    "State":"State",
    "Median_Household_Income_2020":"Median Household Income (2020)",
    "1/31/2020":"Median House Price (1/31/2020)"
})
#rearrange columns
state = merge_df.pop("State")
merge_df.insert(1,"State",state)

#display DataFrame
merge_df
```

```python
#find ratio of median income to median house price
# merge_df.info()
merge_df["Ratio Income to House Price"] = merge_df["Median Household Income (2020)"].astype(float) /  merge_df["Median House Price (1/31/2020)"]
merge_df
```
