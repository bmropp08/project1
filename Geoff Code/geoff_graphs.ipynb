```python
import pandas as pd
import os
import numpy as np
```

```python
#read in the data
df = pd.read_csv('overall_df.csv')
states = pd.read_csv('state_list.csv')
locations = pd.read_csv('uscounties.csv')
```

```python
df.head()
```

```python
states.head()
```

```python
locations.head()
```

```python
#add state to county in locations
locations['county_full'] = locations['county_full'] + ', ' + locations['state_id']

#drop all columns but count_full, lat, and lng
locations = locations[['county_full', 'lat', 'lng']]

#change name of county_full to County in locations
locations = locations.rename(columns={'county_full': 'County'})
#remove the comma in County column
locations['County'] = locations['County'].str.replace(',', '')
#delete the index
locations = locations.reset_index(drop=True)
locations.head()
```

```python
#merge locations to overall
df = pd.merge(df, locations, on='County')
#merge locations to states
states = pd.merge(states, locations, on='County')
df.head()
```

```python
states.head()
```

```python
#rename columns in states
states = states.rename(columns={'Tax Rate': 'Sales Tax', 'salary/house': 'Income to Home Price Ratio', 'property tax rate': 'Property Tax Rate', "percent hs diploma": "Percent of adults with HS Diploma", "overall rank": "Overall Rank" })
```

```python
#graph the states data
import plotly.express as px
fig = px.scatter_mapbox(states, lat="lat", lon="lng", hover_name="County", hover_data=["Sales Tax", "Life Expectancy", "Income to Home Price Ratio", "Percent of adults with HS Diploma", "Overall Rank"], color="Overall Rank", color_continuous_scale=px.colors.cyclical.IceFire, zoom=3)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
#title "Best Counties to Live In by State"

fig.show()
```
